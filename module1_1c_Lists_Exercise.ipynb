## Indexing and Slicing

Access the second item in your mixed types list


```python
print(mixed[1])
```

Access the 4th-7th items in your list of numbers

```python
print(mixed[3:7:])
```

Access the last 4 items in your mixed types list

```python
print(mixed[-4:])
```
